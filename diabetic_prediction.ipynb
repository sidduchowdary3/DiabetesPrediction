```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(13,5))
data_len=diabetic_data[diabetic_data['Outcome']==1]['Glucose'].value_counts()
ax1.hist(data_len,color='red')
ax1.set_title('Having diabetes')
data_len=diabetic_data[diabetic_data['Outcome']==0]['Glucose'].value_counts()
ax2.hist(data_len,color='green')
ax2.set_title('NOT Having diabetes')
fig.suptitle('Glucose Levels')
plt.show()
```

```python
#seeprating the input and output feature
X=diabetic_data.drop(columns='Outcome',axis=1)
Y=diabetic_data['Outcome']
```

```python
#input features
print(X)
```

```python
#output feature
print(Y)
```

```python
#data pre processing or the data standardisation
scaler=StandardScaler()
standard_data=scaler.fit_transform(X)
print(standard_data)
```

```python
#feeding the standard data to the input features
X=standard_data
Y=diabetic_data['Outcome']
print(X)
print(Y)
```

```python
#splitting the data into train and test
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
#training the model
classifier=svm.SVC(kernel='linear') #support vector classifier (svc)
classifier.fit(X_train,Y_train)
```

```python
#finding the accuracy score on the training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy score of the training data :',training_data_accuracy)
```

```python
#finding the accuracy score for the unknown data
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy score of the test data :',test_data_accuracy)
```

```python
#predicting whether the person having the diabetes or not
input_data=(1,103,30,38,83,43.3,0.183,33)

#convert the above data into the numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the array as we are predicting for one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

#standardise the data
std_data=scaler.transform(input_data_reshaped)
prediction=classifier.predict(std_data)
if prediction==0:
  print("Person is not diabetic")
else:
  print("Person is diabetic")
```
